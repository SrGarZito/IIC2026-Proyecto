```python
import pandas as pd

df = pd.read_csv('dataset.csv')
df
```

```python
# Solo Saga's
filtro_namek = df['Saga_or_Movie'].str.contains('Namek', case=False, na=False)
filtro_ginyu = df['Saga_or_Movie'].str.contains('Ginyu', case=False, na=False)
filtro_freezer = df['Saga_or_Movie'].str.contains('Frieza', case=False, na=False)
df_na = df[filtro_namek]
df_gi = df[filtro_ginyu]
df_fr = df[filtro_freezer]
datas = [df_na, df_gi, df_fr]

df = pd.concat(datas)
df
```
