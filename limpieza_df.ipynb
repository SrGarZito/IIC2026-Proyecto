```python
# Solo Saga's
filtro_namek = df[df['Saga_or_Movie'] == 'Namek Saga']
filtro_force = df[df['Saga_or_Movie'] == 'Ginyu Force Saga']
filtro_ginyu = df[df['Saga_or_Movie'] == 'Captain Ginyu Saga']
filtro_freezer = df[df['Saga_or_Movie'] == 'Frieza Saga']

filtro_namek['Power_Level'] = filtro_namek['Power_Level'].replace(',', '', regex=True).astype(float)
filtro_force['Power_Level'] = filtro_force['Power_Level'].replace(',', '', regex=True).astype(float)
filtro_ginyu['Power_Level'] = filtro_ginyu['Power_Level'].replace(',', '', regex=True).astype(float)
filtro_freezer['Power_Level'] = filtro_freezer['Power_Level'].replace(',', '', regex=True).astype(float)

filtro_namek = filtro_namek.sort_values('Power_Level', ascending=False)
filtro_force = filtro_force.sort_values('Power_Level', ascending=False)
filtro_ginyu = filtro_ginyu.sort_values('Power_Level', ascending=False)
filtro_freezer = filtro_freezer.sort_values('Power_Level', ascending=False)

filtro_freezer = filtro_freezer.replace('Dende (potential unlocked)', 'Dende')
filtro_freezer = filtro_freezer.replace('Frieza (100% power)','Freezer')
filtro_freezer = filtro_freezer.replace('Frieza (70% power)','Freezer')
filtro_freezer = filtro_freezer.replace('Frieza (50% power)','Freezer')
filtro_freezer = filtro_freezer.replace("Frieza (cut in half, with Goku's energy)",'Freezer')
filtro_freezer = filtro_freezer.replace("2nd form Frieza (suppressed)",'Freezer')
filtro_freezer = filtro_freezer.replace("Frieza (powered up)",'Freezer')
filtro_freezer = filtro_freezer.replace("2nd form Frieza (full power)",'Freezer')
filtro_freezer = filtro_freezer.replace("Frieza (3rd form)",'Freezer')
filtro_freezer = filtro_freezer.replace("Frieza (10% power, no hands)",'Freezer')
filtro_freezer = filtro_freezer.replace("Freiza (weakening)",'Freezer')
filtro_freezer = filtro_freezer.replace("Frieza's attack",'Freezer')
filtro_freezer = filtro_freezer.replace('Frieza','Freezer')
filtro_freezer = filtro_freezer.replace('Frieza (final form)','Freezer')
filtro_freezer = filtro_freezer.replace('Gohan (angry)','Gohan')
filtro_freezer = filtro_freezer.replace('Gohan (Zenkai)','Gohan')
filtro_freezer = filtro_freezer.replace("Gohan's attack on Frieza",'Gohan')
filtro_freezer = filtro_freezer.replace("Piccolo (Gohan and Krillin's energy)",'Piccolo')
filtro_freezer = filtro_freezer.replace("Piccolo (weighted)",'Piccolo')
filtro_freezer = filtro_freezer.replace("Piccolo (suppressed, after fusion with Nail)",'Piccolo')
filtro_freezer = filtro_freezer.replace("Piccolo (fighting Frieza)",'Piccolo')
filtro_freezer = filtro_freezer.replace("Piccolo (powered up ,weighted)",'Piccolo')
filtro_freezer = filtro_freezer.replace("Piccolo (full power)",'Piccolo')
filtro_freezer = filtro_freezer.replace("Vegeta (rested)",'Vegeta')
filtro_freezer = filtro_freezer.replace("Vegeta (Zenkai)",'Vegeta')
filtro_freezer = filtro_freezer.replace("Vegeta (weakened)",'Vegeta')
filtro_freezer = filtro_freezer.replace("Goku (Zenkai)",'Goku')
filtro_freezer = filtro_freezer.replace("Goku (Kaio-ken x2)",'Goku')
filtro_freezer = filtro_freezer.replace("Goku (Kaio-ken x 3)",'Goku')
filtro_freezer = filtro_freezer.replace("Goku (Kaio-ken x10)",'Goku')
filtro_freezer = filtro_freezer.replace("Goku (Kaio-ken x20)",'Goku')
filtro_freezer = filtro_freezer.replace("Goku's Spirit Bomb",'Goku')
filtro_freezer = filtro_freezer.replace("Goku (SSJ)",'Goku')
filtro_freezer = filtro_freezer.replace("Goku's Angry Kamehameha",'Goku')
filtro_freezer = filtro_freezer.replace("Vegeta's Final Burst Cannon",'Vegeta')

filtro_ginyu = filtro_ginyu.drop([228], axis='index')
filtro_ginyu = filtro_ginyu.replace('Frieza (calm)','Freezer')
filtro_ginyu = filtro_ginyu.replace('Ginyu (full power)','Ginyu')
filtro_ginyu = filtro_ginyu.replace('Vegeta (Zenkai)','Vegeta')
filtro_ginyu = filtro_ginyu.replace('Ginyu (supressed)','Ginyu')
filtro_ginyu = filtro_ginyu.replace('Goku (supressed)','Goku')
filtro_ginyu = filtro_ginyu.replace('Gohan (Zenkai)','Gohan')
filtro_ginyu = filtro_ginyu.replace("Goku (in Ginyu's body)",'Goku')
filtro_ginyu = filtro_ginyu.replace("Goku (Kaio-ken x2)",'Goku')
filtro_ginyu = filtro_ginyu.replace("Ginyu (in Goku's body)",'Ginyu')

filtro_force = filtro_force.replace('Gohan (angry)','Gohan')
filtro_force = filtro_force.replace('Gohan (suppressed)','Gohan')
filtro_force = filtro_force.replace('Krillin (suppressed)','Krillin')
filtro_force = filtro_force.replace('Goku (suppressed)','Goku')
filtro_force = filtro_force.replace('Vegeta (suppressed)','Vegeta')
filtro_force = filtro_force.replace("Vegeta's Final Crash",'Vegeta')
filtro_force = filtro_force.replace("Recoome's Eraser Gun",'Recoome')
filtro_force = filtro_force.replace("Jeice's Crusher Ball",'Jeice')

filtro_namek = filtro_namek.replace('Frieza (calm)','Freezer')
filtro_namek = filtro_namek.replace('Zarbon (transformed)','Zarbon')
filtro_namek = filtro_namek.replace('Nail (suppressed)','Nail')
filtro_namek = filtro_namek.replace('Gohan (potential unlocked)','Gohan')
filtro_namek = filtro_namek.replace('Goku (Zenkai)','Goku')
filtro_namek = filtro_namek.replace('Krillin (potential unlocked)','Krillin')
filtro_namek = filtro_namek.replace('Vegeta (Zenkai)','Vegeta')
filtro_namek = filtro_namek.replace('Namekian Warriors','Namekianos')

filtro_namek = filtro_namek.drop_duplicates(subset='Character', keep='first')
filtro_force = filtro_force.drop_duplicates(subset='Character', keep='first')
filtro_ginyu = filtro_ginyu.drop_duplicates(subset='Character', keep='first')
filtro_freezer = filtro_freezer.drop_duplicates(subset='Character', keep='first')

filtro_freezer = filtro_freezer.drop([233,237,238,236], axis='index')
filtro_namek = filtro_namek.drop([191,185,189], axis='index')

filtro_namek.to_csv('namek.csv')
filtro_force.to_csv('force.csv')
filtro_ginyu.to_csv('ginyu.csv')
filtro_freezer.to_csv('freezer.csv')
```
